I need to find a way of encoding the "booking_origin", "route", "sales_channel" and "trip_type" into the data such that it is meaningful for my RandomForest model similar to the way we have converted the "flight_day" to an integer 1 - 7. 

For "booking origin" and "route" I am going to use frequency encoding since it is straightforward to implement. I will use binary for "sales_channel" (0 = internet , 1 = mobile) and for "trip_type" I will do the same (0 = round trip/ circle trip, 1 = one way).

```python
df['sales_channel'].value_counts()
```

```python
#Encode the countries first
frequency_map_booking_origin = df['booking_origin'].value_counts().to_dict()
df['booking_origin'] = df['booking_origin'].map(frequency_map_booking_origin)

#Encode the flight route
frequency_map_route = df['route'].value_counts().to_dict()
df['route'] = df['route'].map(frequency_map_route)

#Encode the sales_channel
sales_channel_map = {
    "Internet": 0, 
    "Mobile": 1,
}


df["sales_channel"] = df["sales_channel"].map(sales_channel_map)


#Encode the trip type
trip_type_map = {
    "RoundTrip": 0 , 
    "OneWay": 1,
    "CircleTrip": 0,
    
}

df["trip_type"] = df["trip_type"].map(trip_type_map)
```

```python
X = df.drop("booking_complete", axis = 1)
y = df["booking_complete"]
```

```python
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(X,y,train_size = 0.2, random_state = 42)
X_train.shape, X_test.shape
```

```python
from sklearn.ensemble import RandomForestClassifier

classifier_rf = RandomForestClassifier(random_state = 42, n_jobs = -1, max_depth = 5, n_estimators = 100, oob_score = True)


classifier_rf.fit(X_train,y_train)
```

```python
rf = RandomForestClassifier(random_state = 42, n_jobs = -1)

params = {
    "max_depth":[3],
    "min_samples_leaf": [11,12],
    "n_estimators": [43,45,47]
}

from sklearn.model_selection import GridSearchCV

grid_search = GridSearchCV(estimator = rf, param_grid = params, cv=4,n_jobs = -1, verbose = 1, scoring = "accuracy")
```

```python
grid_search.fit(X_train,y_train)
rf_best = grid_search.best_estimator_
rf_best
```

```python
from sklearn.tree import plot_tree
import matplotlib.pyplot as plt

plt.figure(figsize = (150,100))
plot_tree(rf_best.estimators_[5], feature_names = list(X.columns.values), class_names = ["Won't book", "Will book"], filled = True)
```
